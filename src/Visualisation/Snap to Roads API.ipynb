#### Get points within polygons

```python
import requests
import random
import shapefile
from shapely.geometry import Polygon, Point
from shapely import wkt

# Read shapefile
r = shapefile.Reader(r"D:\FYP\Visualisation\QGIS\KL_Neighbourhood.shp")
sr = r.shapeRecords()

# Create Shapely Polygons
polygons = []

for point in sr:
    polygons.append([Polygon(point.shape.points) , point.record[0]])

def createPoints(points, numberOfPoints, polygon, latitudeBoundaries, longitudeBoundaries):
    for i in range(numberOfPoints):
        point = (Point(random.uniform(longitudeBoundaries[0],longitudeBoundaries[1]), random.uniform(latitudeBoundaries[0],latitudeBoundaries[1])))
        
        if (point.within(polygon)):
            # for unique points
            point = wkt.dumps(point)
            point = (point.replace('POINT (',''))
            point = (point.replace(')',''))
            points.append(point)
            if (len(set(points)) == numberOfPoints):
                break
        else:
            pass
    return points
        
randomPoints = []
numberOfPoints = 20
# Create random Shapely points in bounding boxes for each of the polygons
for polygon in (x[0] for x in polygons):
    print("Next")
    # create bounding box of the polygon
    # polygon.bounds returns (minx, miny, maxx, maxy)
    longitudeBoundaries = [polygon.bounds[0],polygon.bounds[2]]
    latitudeBoundaries = [polygon.bounds[1], polygon.bounds[3]]  
    
    points = []
#     count = 0
    while (len(set(points))<numberOfPoints):
        points = createPoints(points, numberOfPoints, polygon, latitudeBoundaries, longitudeBoundaries)
        print(len(points))
#         count+=1
    # change back to coordinate form before appending

    points = [point.split(" ") for point in points]
    randomPoints.append(points)
print(len(randomPoints))
```

#### Snap to Roads Request

```python
df = []
numberOfPoints = 20
# loop over the polygons
for polygon, neighbourhoodID in polygons:
    # loop over the points in the polygon
    for j in range(numberOfPoints):
        url = "https://roads.googleapis.com/v1/snapToRoads?path={},{}&interpolate=true&key=AIzaSyBk5oVZc2neQcMBWD4Yr02JeQkbjETzUJk".format(randomPoints[neighbourhoodID][j][1],randomPoints[neighbourhoodID][j][0])
        
        x = requests.get(url)
        
        x = x.json()
        try:
            latitude = (x["snappedPoints"][0]["location"]["latitude"])
            print(latitude)
            longitude = (x["snappedPoints"][0]["location"]["longitude"])
            print(longitude)
            panoID = (x["snappedPoints"][0]["placeId"])
            print(panoID)
            print(neighbourhoodID)

            df.append([panoID, latitude, longitude, neighbourhoodID])
        except:
            pass
        
#         print(x.text)
```

```python
import pandas as pd

data = pd.DataFrame(df)
data.columns = ["panoID", "latitude", "longitude", "neighbourhoodID"]

data.to_csv(r"D:\FYP\Visualisation\test_points_KL.csv")
```
